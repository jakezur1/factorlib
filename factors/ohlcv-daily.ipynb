```python
import pandas as pd
```

```python
ohlcv_data = pd.read_csv('../data/ohlcv-wrds-daily.csv')
ohlcv_data['datadate'] = pd.to_datetime(ohlcv_data['datadate'])
ohlcv_data.drop(columns=['tic'], inplace=True)
ohlcv_data.set_index(['GVKEY', 'datadate'], inplace=True)
```

```python
ohlcv_by_ticker = ohlcv_data.unstack(level=0)
ohlcv_by_ticker = ohlcv_by_ticker.swaplevel(0, 1, axis=1)
ohlcv_by_ticker.sort_index(axis=1, level=0, inplace=True)
```

```python
ohlcv_by_ticker.to_csv('../data/ohlcv-daily-by-ticker.csv')
```
